## Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,x**3)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x**2,y)
```

```python
for i in range(1,37,1):  # There will be only one canvas and all the subplots will be in that canvas sheet
    plt.subplot(6,6,i)
    if i==35:
        plt.plot(x,y)
```

## Object Oriented Way

##### Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

##### figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
# The white background is known as canvas
```

```python
fig = plt.figure()
# fig # It is just a canvas sheet
fig.add_axes([0.1,0.1,0.6,0.8]) # left bottom width height
```

```python
fig = plt.figure()
fig.add_axes([0.1,0.1,0.3,0.5]) # left bottom width height
```

```python
fig = plt.figure()
fig.add_axes([0.1,0.1,1,1]) # always provide value btw 0 and 1
# Full canva as height and width is 100%
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.5,0.3,0.3])
axis2.plot(x,y)
```

```python
fig = plt.figure(figsize=(12,8),dpi=100) # figsize=(width,height) and dpi = 100
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.5,0.3,0.3])
axis2.plot(x,y)
```
