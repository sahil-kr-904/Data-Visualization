```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### We use distribution plots to visualize the distribution of quantitative data.

##### Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

- distplot/histplot
- jointplot
- pairplot
- rugplot

```python
df = sns.load_dataset("tips")
```

```python
df 
# Quantitative Variable -> multiple various values that can't be categorized (Age,Salary)
# Qualitative Variable -> values that can be categorized(Gender)
# Distribution Plots are created for quantitative variable
```

```python
# total_bill -> Qunatitative
# tip -> Quantitative
# Sex,Smoker,day,time,size -> Qualitative Variable or categorical data
```

```python
df["size"].unique()
```

```python
sns.displot(df["total_bill"]) # histogram is best in showing distribution # bin
```

```python
sns.displot(df["total_bill"],bins=1)
```

```python
sns.displot(df["total_bill"],bins=10)
```

```python
sns.displot(df["total_bill"],bins=100)
```

```python
sns.histplot(df["total_bill"],bins=10) # Both is same
```

```python
sns.displot(df["total_bill"],bins=20,kde=True)
```

```python
plt.subplot(2,2,1)
sns.histplot(df["total_bill"],bins=20,kde=True) # use histplot
plt.subplot(2,2,4)
sns.histplot(df["tip"],bins=20,kde=True)
```

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind="scatter")
```

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind="kde")
```

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind="hex")
```

```python
sns.jointplot(x="total_bill",y="tip",data=df,kind="reg")
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex") # by using hue, we separate out data in category wise
```

```python
sns.pairplot(df,hue="sex",palette="rainbow") # designing
```

```python
sns.pairplot(df,hue="size",palette="rainbow")
```

```python
sns.rugplot(df["total_bill"])
```

```python
sns.rugplot(df)
```
